```python
david_img_green=np.zeros((320,240,3)) # creating a array of image size with zero
david_img_green[:,:,1]=david_img[:,:,1] # converting green channel image of original image
plt.imshow(david_img_green) # green channel image
```

```python
david_img_blue=np.zeros((320,240,3)) # creating a array of image size with zero
david_img_blue[:,:,2]=david_img[:,:,2] # converting blue channel image of original image
plt.imshow(david_img_blue) # blue channel image
```

```python
print(david_img[0,0,:])
print(david_img[6,5,:])
```

```python
david_img_dif=2*david_img[:,:,0]-david_img[:,:,1]-david_img[:,:,2]
print(david_img_dif) # deducting gray scale image matrix
```

```python
plt.imshow(david_img_dif, cmap='gray') # the new gray scale image
```

```python
edge_filter_kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]]) # loading the edge fliter kernel 
edges_filter = scipy.signal.convolve2d(david_img_dif, edge_filter_kernel, 'valid') # appplying the kernel on the gray scal image
```

```python
plt.imshow(edges_filter, cmap=plt.cm.gray) # display the image
```
